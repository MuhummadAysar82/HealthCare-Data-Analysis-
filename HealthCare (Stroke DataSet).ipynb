```python
plt.bar(df['gender'].value_counts().index, df['gender'].value_counts().values, color= ['skyblue', 'pink', 'purple'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize= (7,7))
plt.pie(df['work_type'].value_counts().values, labels= df['work_type'].value_counts().index,autopct= '%1.1f%%', startangle=90, colors= ['skyblue', 'lightgreen', 'orange', 'pink', 'violet'])
plt.title('Work Type Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Age Distribution Of People')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis= 'y', linestyle= '--', alpha= 0.7)
plt.show()
```

```python
plt.figure(figsize= (7,7))
plt.pie(df['smoking_status'].value_counts().values, labels = df['smoking_status'].value_counts().index, autopct= '%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'orange', 'pink'])
plt.title('Smoking Status Distribution')
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df[df['gender'] == 'Male']['heart_disease'], bins=[-0.5, 0.5, 1.5], alpha=0.7, label='Male', color='skyblue')
plt.hist(df[df['gender'] == 'Female']['heart_disease'], bins=[-0.5, 0.5, 1.5], alpha=0.7, label='Female', color='pink')

plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])
plt.title('Heart Disease Distribution by Gender')
plt.xlabel('Heart Disease Status')
plt.ylabel('Number of People')
plt.legend()
plt.show()
```
